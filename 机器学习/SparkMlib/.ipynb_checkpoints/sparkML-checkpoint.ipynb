```python
"""
fbprophet
"""
from fbprophet import Prophet
import pandas as pd
import numpy as np
path = '/home/hadoop/data/sparkMlib/TotalFee.csv'
df = pd.read_csv(path)
df.head()
```

```python
df.columns = ['ds', 'y']
df.head()
```

```python
m = Prophet()
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
m.plot(forecast)
```

```python
m.plot_components(forecast)
```
